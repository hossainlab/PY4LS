## Distribution Plot
The distplot shows the distribution of a univariate set of observations.
- Documentation: https://seaborn.pydata.org/tutorial/distributions.html

### Univariate Distribution - Histograms
### Histograms
4 Main Aspects

- Shape: Overall appearance of histogram. Can be symmetric, bell-shaped, left skewed, right skewed, etc.

- Center: Mean or Median

- Spread: How far our data spreads. Range, Interquartile Range (IQR),standard deviation, variance.

- Outliers: Data points that fall far from the bulk of the data
- Documentation: https://seaborn.pydata.org/generated/seaborn.histplot.html

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=2, palette= "viridis")
sns.histplot(data=df, x="Age")
# plt.savefig("distribution_age.png", dpi=300)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=2, palette= "viridis")
sns.histplot(data=df, y="Age")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=2, palette= "viridis")
sns.histplot(data=df, x="Age", bins=20)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=2, palette= "viridis")
sns.histplot(data=df, x="Age", bins=20, binwidth=3)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=2, palette= "viridis")
sns.histplot(data=df, x="Age", hue='Gender')
plt.show()
```

## Bivariate Distribution - Jointplot
jointplot() allows you to basically match up two distplots for bivariate data. With your choice of what kind parameter to compare with:

- scatter

- reg

- resid

- kde

- hex


```python
sns.set(font_scale=2, palette= "viridis")
sns.jointplot(x='Age',y='LungCap',data=df,kind='scatter')
plt.show()
```

```python
sns.set(font_scale=2, palette= "viridis")
sns.jointplot(x='Age',y='LungCap',data=df,kind='hex')
plt.show()
```

```python
sns.set(font_scale=2, palette= "viridis")
sns.jointplot(x='Age',y='LungCap',data=df,kind='reg')
plt.show()
```

## Matrixplots / Heatmaps

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("../data/LungCapData.xls")
```

```python
df.head()
```

```python
pd.crosstab(df.Gender, df.Smoke)
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.4)
sns.heatmap(pd.crosstab(df.Gender, df.Smoke),annot= True,  fmt = "d", cmap = "Reds", vmax = 150)
plt.show()
```

```python
pd.crosstab(df.Gender, df.Smoke, values= df.LungCap, aggfunc= "mean")
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.4)
sns.heatmap(pd.crosstab(df.Gender, df.Smoke, values= df.LungCap, aggfunc= "mean"), annot= True, cmap = "Reds")
plt.show()
```

```python
df.corr()
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.4)
sns.heatmap(df.corr(), annot= True, cmap = "Reds")
plt.show()
```

## Pair Plot
Plot pairwise relationships in a dataset.

- By default, this function will create a grid of Axes such that each numeric variable in data will by shared across the y-axes across a single row and the x-axes across a single column. The diagonal plots are treated differently: a univariate distribution plot is drawn to show the marginal distribution of the data in each column.
- It is also possible to show a subset of variables or plot different variables on the rows and columns

```python
import pandas as pd 
import seaborn as sns 
```

```python
df = pd.read_csv('../data/diabetes.csv')
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='Outcome')
```
